```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
from datetime import datetime
import re
```

```python
genre = ['drama', 'crime', 'comedy', 'thriller', 'action', 'romance', 'SF', 'horror', 'fantasy']

drama_url = ['https://www.imdb.com/title/tt10919420/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt14169960/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt12235718/reviews?ref_=tt_urv']
crime_url = ['https://www.imdb.com/title/tt12235718/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt13433812/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt13885336/reviews?ref_=tt_urv']
comedy_url = ['https://www.imdb.com/title/tt15026724/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt13433812/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt11769304/reviews?ref_=tt_urv']
thriller_url = ['https://www.imdb.com/title/tt10919420/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt14169960/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt12235718/reviews?ref_=tt_urv']
action_url = ['https://www.imdb.com/title/tt10919420/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt14169960/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt13433812/reviews?ref_=tt_urv']
romance_url = ['https://www.imdb.com/title/tt17513352/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt11239552/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt15181094/reviews?ref_=tt_urv']
SF_url = ['https://www.imdb.com/title/tt14169960/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt12235718/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt11612120/reviews?ref_=tt_urv']
horror_url = ['https://www.imdb.com/title/tt14169960/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt12235718/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt11570202/reviews?ref_=tt_urv']
fantasy_url = ['https://www.imdb.com/title/tt14169960/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt12235718/reviews?ref_=tt_urv', 'https://www.imdb.com/title/tt13273826/reviews?ref_=tt_urv']

WHOLE_URL = [drama_url, crime_url, comedy_url, thriller_url, action_url, romance_url, SF_url, horror_url, fantasy_url]
review = dict()
```

```python
driver = webdriver.Chrome('/opt/homebrew/Caskroom/chromedriver/99.0.4844.51/chromedriver')
driver.set_page_load_timeout("20")

for i in range(len(WHOLE_URL)):
    for j in range(3):
        url = WHOLE_URL[i][j]
        time.sleep(5)
        driver.get(url)
        
        # 작품 제목
        title = driver.find_element_by_css_selector('#main > section > div.subpage_title_block > div > div > h3 > a').text
        print("title", title)
        
        # 댓글 수
        review_counts = driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[3]/div[1]/section/div[2]/div[1]/div/span').text
        review_counts = int(review_counts.split(' ')[0].replace(',', ''))
        print(f"reviews: {review_counts}")
        
        # Load More 끝날 때까지 반복
        for k in range(review_counts):    
            try:
                driver.find_element_by_class_name("ipl-load-more__button").send_keys(Keys.ENTER)
                time.sleep(3)
            except Exception as ex:
                print('더이상 "Load More" 없음', ex)
                break

        print(" 댓글 수집 ")
        try:
            time.sleep(0.5)
            soup= BeautifulSoup(driver.page_source)
            time.sleep(1)

            comment=[]
            x= soup.find_all("div", class_="content")

            for h in x:
                try:
                    tmp= h.find_all("div", class_="text show-more__control")[0].text
                    comment.append(tmp)
                except Exception as ex:
                    tmp= h.find_all("div", class_="text show-more__control clickable")[0].text
                    comment.append(tmp)

            time.sleep(1)
        except Exception as ex:
            re=['nothing']
            print("댓글없음")

        print("댓글 평점 수집 ")
        try:
            rating=[]
            x= soup.find_all("div", class_="lister-item-content")
            for h in x:
                try:    
                    a1= h.find_all("span", class_="rating-other-user-rating")[0]
                    a2= a1.find_all("span")[0].text
                    rating.append(a2)
                except Exception as ex:
                    rating.append('99')

            time.sleep(0.5)

        except Exception as ex:
            rating= 'nothing'
            print("댓글 평점 없음")

        time.sleep(1)
        print(datetime.now())
        print("---제목, 수집 댓글, 평점 저장---")
        review[title] = [comment,rating,genre[i]]
```
