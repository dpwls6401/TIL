```python
import numpy as np
import pandas as pd

import seaborn as sns
sns.__version__
```

```python
flights=sns.load_dataset('flights')
flights
```

```python
#relplot; relational plot
#뿌연 구간; 95% 신뢰구간
sns.relplot(x='year',y='passengers',kind='line',ci='sd',data=flights);
```

```python
sns.relplot(x='year',y='passengers',kind='line',estimator=None,data=flights);
```

```python
dots=sns.load_dataset('dots')
dots
```

```python
#hue: color, style; 모양
sns.relplot(x='time',y='firing_rate',kind='line',hue='align',style='choice',markers=True,data=dots);
```

```python
sns.relplot(x='time',y='firing_rate',kind='line',hue='align',row='choice',data=dots);
```

```python
sns.relplot(x='time',y='firing_rate',kind='line',hue='align',col='choice',data=dots);
```

```python
sns.relplot(x='time',y='firing_rate',kind='line',hue='coherence', style='choice',data=dots[dots['align']=='sacc']);
```

```python
dots.query("align=='sacc'") #dots[dots['align']=='sacc']와 동일
```

```python
fmri=sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri);
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',hue='subject',style='region',size='event',data=fmri);
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',hue='subject',col='region',size='event',data=fmri);
```

```python
#cubehelix_palette; 색깔 변경할 때 이용
palette=sns.cubehelix_palette(n_colors=14, light=0.8)
sns.relplot(x='timepoint',y='signal',kind='line',hue='subject',col='region',size='event',palette=palette, data=fmri);
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',hue='subject',col='region',row='event',palette=palette, data=fmri);
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',hue='event',col='subject',linewidth=3,col_wrap=4,data=fmri);
```

```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
sns.catplot(x='species',y='body_mass_g',kind='box',data=penguins);
```

```python
sns.catplot(x='species',y='body_mass_g',kind='box',hue='sex',data=penguins);
```

```python
sns.catplot(x='species',y='body_mass_g',kind='box',col='sex',data=penguins);
```

```python
#pivot(index, col, val)
#col; array로 받음
flights_pivot=flights.pivot('month','year','passengers')
flights_pivot
```

```python
sns.heatmap(flights_pivot);
```

```python
#annot; 값 보여줌
sns.heatmap(flights_pivot, annot=True,fmt='d');
```

```python
sns.heatmap(flights_pivot, cmap='BuPu');
```

```python
sns.heatmap(flights_pivot, cmap='BuPu',cbar=False);
```
