```python
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", header=None)
df
```

```python
df.columns=['Class label','Alcohol','Malic acid', 'Ash', 'Alcalinity of ash','Magnesium','Total phenols',
           'Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines',
           'Proline']
print("클래스 레이블", np.unique(df['Class label']))
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.iloc[:,1:].values
y=df.iloc[:,0].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0,stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
stdsc=StandardScaler()
X_train_std=stdsc.fit_transform(X_train)
X_test_std=stdsc.transform(X_test)

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(solver='liblinear', penalty='l1',C=1.0, random_state=1)
lr.fit(X_train_std,y_train)
print("훈련 정확도:", lr.score(X_train_std,y_train))
print("테스트 정확도:",lr.score(X_test_std,y_test))
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
ax=plt.subplot(111)

colors=['blue','green','red','cyan','magenta','yellow','black','pink','lightgreen','lightblue','gray','indigo','orange']
weights,params=[],[]

for c in np.arange(-4.,6.):
    lr=LogisticRegression(solver='liblinear', penalty='l1',C=1.0, random_state=1)
    lr.fit(X_train_std,y_train)
    
    weights.append(lr.coef_[1])
    params.append(10**c)

weights=np.array(weights)

for column,color in zip(range(weights.shape[1]),colors):
    plt.plot(params,weights[:,column], label=df.columns[column+1], color=color)
    
plt.axhline(0,color='black',linestyle='--',linewidth=3)
plt.xlim([10**(-5), 10**5])
plt.ylim(-25,10)
plt.xlabel('C')
plt.ylabel('weight coefficient')
plt.xscale('log')
plt.legend(loc='upper left')
ax.legend(loc='upper center', bbox_to_anchor=(1.38,1.03), ncol=1, fancybox=True)
plt.show();   
```
