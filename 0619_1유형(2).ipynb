2. 결측치 확인하기

```python
#라이브러리, 데이터 로드
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/bigData-main/boston.csv')

#결측치 확인
# print(data.isnull().sum())

#평균값 대치
data_mean = data['RM'].copy()
rm_mean = data['RM'].mean()
data_mean.fillna(rm_mean, inplace=True)
# print(data_mean.isnull().sum())

#결측치 삭제
data_del = data['RM'].copy()
data_del.dropna(inplace=True)
# print(data_del.isnull().sum())

#표준편차 값 차이, 절대값
print(abs(data_mean.std() - data_del.std()))
```

3. 이상값 확인하기

```python
#라이브러리, 데이터 로드
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/bigData-main/boston.csv')

#최대경계값, 최소경계값
zn = data['ZN'].copy()
zn_mean = zn.mean()
zn_std = zn.std()

zn_max = zn_mean + (1.5 * zn_std)
zn_min = zn_mean - (1.5 * zn_std)

#이상값 합계
print(sum(zn[zn > zn_max]) + sum(zn[zn < zn_min]))
```

4. 사분위수 구하기

```python
#라이브러리, 데이터 로드
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/bigData-main/boston.csv')

#칼럼 제외
data_new = data.drop(columns=['CHAS', 'RAD'])
# print(data_new.head())

#칼럼별 IQR
data_desc = data_new.describe()
# print(data_desc)

data_new = data_desc.iloc[[4,6]].T
# print(data_new)

data_new['IQR'] = data_new['75%'] - data_new['25%']
print(data_new['IQR'])
```

5. 순위 구하기

```python
#라이브러리, 데이터 로드
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/bigData-main/boston.csv')

#내림차순 정렬
data_new = data.sort_values(by='MEDV', ascending=False)

#30번째로 큰 값 적용
# data_new['MEDV'].head(30)
data_new['MEDV'][0:29] = 41.7
# data_new['MEDV'].head(30)

#평균값, 중위값, 최솟값, 최댓값
print(data_new['MEDV'].mean(), data_new['MEDV'].median(), data_new['MEDV'].min(), data_new['MEDV'].max())
```
